### Importing Libraries

```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem import LancasterStemmer, PorterStemmer

#stemming(어간추출) 방식 중에서 porter 방식을 사용
```

### Data Cleaning and data preprocessing

```python
# 어간 추출기 (PorterStemmer())와 말뭉치 리스트 corpus 생성
ps = PorterStemmer()
ls = LancasterStemmer()
corpus = []
```

```python
df[['message']]
```

```python
df[['target']]
```

```python
for i in range(len(df)):
    review = re.sub('[^a-zA-z]', ' ', df['message'][i])
    review = review.lower()
    review = review.split()
    print(review)
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

'message' 데이터 전처리 과정.
re.sub ==> re.sub('해당하는 정규표현', '대체하고자 문자열', '대상')
1) review = re.sub('[^a-zA-z]', ' ', df['message'][i] --> 영어가 아닌 문자열을 공백으로 치환한다.

2) review = review.lower()
   review = review.split()
   --> 모든 문자열을 소문자로 변환시키고, 공백(띄어쓰기) 기준으로 split

3) review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
--> review에 있는 단어가 불용어가 아니라면, 이 단어를 PorterStemmer방식으로 어간 추출화

4) review = ' '.join(review)
   corpus.append(review)
   --> 어간추출화한 단어들을 다시 문장(말뭉치)으로 이어주고, 말뭉치 리스트에 추가

```python
for cor in corpus:
    print(cor)

len(corpus)
```

### Converting data into Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vect = CountVectorizer()
X = vect.fit_transform(corpus).toarray()
X.shape
```

```python
df['target'].replace(['ham', 'spam'], [0, 1], inplace=True)
df['target']
```

```python
y = df['target']
y
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.6, random_state=5)
```

### Training The Model

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

### Measuring The Accuracy

```python
from sklearn import metrics
```

```python
print("accuracy: ", metrics.accuracy_score(y_test, y_pred=y_pred))
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
print(cm)
```
